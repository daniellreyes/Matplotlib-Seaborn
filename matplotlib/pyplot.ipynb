```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creacion de valores para x y y -> creacion de una parábola
x = np.linspace(0,5,11)
y = x**2
```

Plot es un grafico de líneas -> .plot

```python
plt.plot(x,y, 'rx',)
plt.show()
```

```python
plt.plot(x,y, 'yD:')
plt.show()
```

Para graficar un histograma, se agrega -> .hist

```python
plt.hist(x)
plt.show()
```

Para el gráfico de pastel -> .pie

```python
plt.pie(x)
plt.show()
```

Para la relacion de dos variables -> scatter

```python
plt.scatter(x,y)
plt.show()
```

Para hacer un boxplot -> .boxplot

```python
plt.boxplot(x)
plt.show()
```
