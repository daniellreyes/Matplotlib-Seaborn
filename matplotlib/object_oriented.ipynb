```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Notación orientado a objetos

```python
#Esto es el lienzo -> figure
#Los dos primeros puntos hacen referencia a la posición de la figura en el lienzo y los otros dos al tamaño de los ejes
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.2,2])
axes.plot(x,y,'b')
fig.show()
```

```python
#Esto es el lienzo -> figure
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.1,0.1,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
fig.show()
```

```python
#Esto es el lienzo -> figure
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes.plot(x,y,'b')
axes2.plot(y,x,'r')
axes2.set_facecolor('gray')
fig.show()
```
