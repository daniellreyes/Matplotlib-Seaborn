```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

```python
#Esto es con el método orientado a objetos
fig, axes = plt.subplots(1,2, figsize=(8,5))
axes[0].plot(x,y, label='$sin(x)$')
axes[0].set_title('Función sen(x)')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].legend()

axes[1].plot(y,x)
axes[1].set_title('Función inversa de sen(x)')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
```

```python
plt.plot(x,y,label='sin(x)')
plt.title('Titulo de Sin(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper right', bbox_to_anchor=(1,0.9))
```
