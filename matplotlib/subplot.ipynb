```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

Cómo le hago para que muestre dos gráficas?
1 -> Numero de filas
2 -> Numero de columnas
3 -> index entre filasxcolumnas o dicho de otra forma, posición del arreglo

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, 'r--')
plt.subplot(2,1,2)
plt.hist(y)
plt.show()
```
