```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
country = ['Mexico', 'Japon','Italia', 'Alemania']
population = [1000,699,500,740]
```

```python
plt.bar(country, population, width=0.5, color=['orange','green'])
plt.xticks(np.arange(4),('Mexico','Japon','Italia','Alemania'), rotation=45)
plt.show()
```

```python
#Para usar un formato horizontal, basta con agregar .barh
plt.barh(country, population)
plt.show()
```
