```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = np.sin(x)
```

Definiendo los objetos

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
#Recordar que el primer parmetro es el renglon y el otro la columnna
fig, axes = plt.subplots(2, 4)
axes[0,0].plot(x,np.cos(x),'b')
axes[0,1].plot(y,x,'r')
axes [1,3].plot(x,np.tan(x),'y')
```

```python
#Definiendo previamente la salida del arreglo
fig, ((ax1,ax2,ax3,ax4),(ax5,ax6,ax7,ax8)) = plt.subplots(2, 4)
ax1.plot(x,np.cos(x),'b')
ax2.plot(y,x,'r')
ax7.plot(x,np.tan(x),'y')
```
